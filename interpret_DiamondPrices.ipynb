```python
# Verimizi düzenleyeceğimiz pandas kütüphanemizi ekliyoruz.
import pandas as pd

# Verimizi Train-Test olarak ayıracağımız sklearn kütüphanemizi ekliyoruz.
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Ağaç tabanlı algoritmalarımızı sklearn kütüphanemizden ekliyoruz.
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# XAI algoritmalarımızdan interpret algoritmamızı ekliyoruz.
from interpret.glassbox import ExplainableBoostingClassifier
from interpret import show
```

Verimizi " df " değişkenine atıyoruz

```python
df = pd.read_csv("veriler/DiamondsPrices.csv")
df1 = df.drop("cut",axis=1)
df
```

Verimizi düzenliyoruz

```python
df1 = pd.get_dummies(df1)
df1["cut"] = df["cut"]
df1
```

Verimizin içindeki "cut" kolonunda hangi veriden kaç tane olduğuna bakıyoruz.

```python
count_class = df1.groupby("cut").size()
count_class
```

Train ve Test verilerimizi ayırıyoruz.

```python
X = df1.drop("cut", axis=1)
Y = df1["cut"]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

Modelimizi eğitiyoruz.

```python
ebm = ExplainableBoostingClassifier()
ebm.fit(x_train, y_train)
```

Modelimizin doğruluğuna bakıyoruz.

```python
y_predicted = ebm.predict(x_test)
accuracy = accuracy_score(y_test, y_predicted)
print("Precisión: ",  accuracy*100, "%")
```

Modelimizi daha rahat anlayabilmemiz için İnterpret kütüphanesinin içindeki show() fonksiyonunu kullanıyoruz.

```python
ebm_global = ebm.explain_global()
show(ebm_global)
```

Tahminleri daha iyi anlayabilmemiz için tekrardan show() fonksiyonunu kullanıyoruz

```python
ebm_local = ebm.explain_local(x_test, y_test)
show(ebm_local)
```
