```python
# Verimizi işleyeceğimiz kütüphaneleri ekliyoruz.
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Makine öğrenmesi kütüphanelerimizi ekliyoruz.
from sklearn.tree import DecisionTreeClassifier
# XAI kütüphanemizi dahil ediyoruz.
import eli5
```

```python
# Modeli oluşturmak için verilerimizi çekiyoruz.
names = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
input_names = names[:len(names)-1]
df = read_csv("veriler/Iris.csv")
```

```python
# x değişkenimize "Id" ve "Species" kolonlarının dışındaki verilerimizi atıyoruz.
x = df.drop(["Species","Id"], axis=1)
# y değişkenimize "Species" kolonumuzdaki değişkenleri atıyoruz
y = df["Species"]
# x ve y verilerimizi train ve test olarak ayırıyoruz. (Test için ayırdığımız veri büyüklüğü %33)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
# Decision Tree Classifier (Karar Ağacı Sınıflandırıcı) algoritmamızı eğitiyoruz.
dt_model = DecisionTreeClassifier().fit(x_train, y_train)
```

Decision Tree Classifier, bir makine öğrenimi yöntemidir ve verilen veri kümesine uygulanarak bir sınıflandırma modeli oluşturur. Bu yöntem, veri kümesindeki özelliklerin sınıflandırma tahminlerinde ne kadar etkili olduğunu değerlendirir ve bu bilgilere göre bir karar ağacı oluşturur. Bu karar ağacı, bir örnek üzerinde sınıflandırma tahmini yaparken hangi özelliklerin kullanılacağını ve hangi sınıfa ait olduğunu gösterir. Decision Tree Classifier yöntemi, veri kümesinin özelliklerinin sayısı az ise iyi performans gösterir ancak veri kümesinin özelliklerinin sayısı çok ise overfitting (aşırı uyumluluk) problemine neden olabilir. Bu nedenle, Decision Tree Classifier yöntemini kullanırken veri kümesinin özelliklerinin sayısına dikkat etmek gerekir.

```python
# Doğruluk değerimizi ölçüyoruz
y_predicted = dt_model.predict(x_test)
accuracy = accuracy_score(y_test, y_predicted)
print("Precisión: ",  accuracy*100, "%")
```

eli5.show_weights() fonksiyonu, verilen bir sınıflandırıcı modelin ağırlıklarını gösterir. Bu ağırlıklar, modelin her bir özelliğin ne kadar etkili olduğunu gösterir.

```python
eli5.show_weights(dt_model, feature_names=input_names)
```

Ağırlık değerleri, decision tree modelinin hangi özelliklerin daha önemli olduğunu gösterir. Örneğin, "Weight" başlığı altında yer alan değerler, modelin özellikler arasındaki önem sırasını gösterir. Bu veri setinde, PetalWidthCm özellği en önemli özellik olarak kabul edilir ve ağırlık değeri 0.5446'dır.

Daha sonra, "Feature" başlığı altında yer alan değerler, modelin karar verme sürecinde kullandığı kuralları gösterir. Örneğin:

"PetalWidthCm <= 0.800 (34.0%) ---> [1.000, 0.000, 0.000]" ifadesi, modelin PetalWidthCm özelliğinin 0.800'den küçük olan veri noktaları için, species sınıfı olarak 1.000 değerini, diğer iki sınıf için de 0.000 değerini verme kuralını gösterir. Bu kural, PetalWidthCm özelliğinin 0.800'den küçük olan veri noktaları için %34 olasılıkla geçerlidir.

Bu şekilde, modelin özellikler arasındaki önem sırası ve karar verme sürecinde kullandığı kurallar gösterilmiş olur. Bu bilgiler, modelin nasıl çalıştığını anlamaya yardımcı olur.





"show_prediction()" fonksiyonu, verilen bir örnek üzerinde sınıflandırma tahmini yaparken modelin nasıl çalıştığını açıklayan bir özet sunar. Bu fonksiyon, verilen örnek üzerinde sınıflandırma tahmini yaparken modelin hangi özellikleri kullandığını, bu özelliklerin tahminler üzerindeki etkilerini ve tahminin hangi sınıfa ait olduğunu gösterir.

Burada 0. ve 10. test verisini kullanmışız

```python
i = 0
eli5.show_prediction(dt_model, x_test.iloc[i], feature_names=input_names, show_feature_values=True)
```

Tamam, bu kısımda verilen çıktıda, decision tree modelinin tahmin yapma sürecinde kullandığı kurallar gösterilmektedir. Bu çıktıda, yapılan tahminlerin hangi sınıfa ait olduğu ve bu tahminleri etkileyen özelliklerin ağırlık değerleri gösterilmektedir.

y=Iris-versicolor (probability 1.000) başlığı altında yer alan çıktıda, decision tree modelinin, veri noktasının Iris-versicolor sınıfına ait olduğu tahminini gösterir. Bu tahmin sonucuna göre, veri noktasının probability (olasılık) değeri 1.000'dir. Bu çıktının altında yer alan "Contribution?" başlığı altında yer alan değerler ise, tahmin sonucunu etkileyen özelliklerin ağırlık değerlerini gösterir. Örneğin, "Feature" başlığı altında yer alan PetalLengthCm özelliğinin ağırlık değeri 0.455'dir ve bu özellik tahmin sonucunu etkilemiştir.
